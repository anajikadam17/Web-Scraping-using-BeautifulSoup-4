# Web Scraping using BeautifulSoup 4

## Web scrapping for Arabic Reviews: Practice Task


```python
import re
import emoji
import pandas as pd
import numpy as np

from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

## For single URL

```python
def project_reviews(url):
    title_1 = []
    Reviews_1 = []
    Ratings_1 = []
    for i in range(1,60):  # for page number update 1 to 60
        j = i
        url = url+str(i)
        #print(url)
        html = urlopen(url)
        soup = BeautifulSoup(html)
        #print('Page:',i)
        title =soup.findAll("a", {"class" : "a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"})
        for i in title:
            a = emoji.get_emoji_regexp().sub(u'', i.text)         # remove emoji
            ab = re.search('[a-zA-Z]',a)
            if ab:
                title_1.append(a[1:-1])
            else:
                title_1.append(a[1:-1])
        Reviews =soup.findAll("span", {"class" : "a-size-base review-text review-text-content"})
        for i in Reviews:
            a = emoji.get_emoji_regexp().sub(u'', i.text)         # remove emoji
            ab = re.search('[A-Za-z0-9]',a)                       # check for english review
            if ab:
                Reviews_1.append('English_Reviews')               # English review tag
            else:
                Reviews_1.append(a[2:-2])                         # other than english
        Ratings = soup.findAll("span", {"class" : "a-icon-alt"})  # for rating
        for i in range(3,len(Ratings)):
            Ratings_1.append(Ratings[i].text[:3])                 # save ratings
        print('__Page No.'+str(j)+'_____'*10)
    dict1 = {'rating':Ratings_1, 'Title':title_1, 'Review':Reviews_1}  # create dictionary
    df = pd.DataFrame(dict1)                                           # create DataFrame
    #print(df.shape)
    return df

# URL to scrap-->
url = ["https://www.amazon.sa/Apple-AirPods-Charging-Case-White/product-reviews/B07NWBVQWK/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber="
       ]

df = pd.DataFrame(columns=['rating', 'Title', 'Review'])
for i in url:
    df1 = project_reviews(i)
    df = df.append(df1,ignore_index=True)
df.shape
df.to_csv('file.csv')
```

##  Add CSV files and create a master CSV File

```python
# Check Folders items
import os 
import os
import re
import numpy as np
import pandas as pd
for file in os.listdir():
    csv = [ file for file in os.listdir() if file.endswith('.csv') ] # Check .csv file in folder and create list

print('There are '+str(len(csv))+' csv files :',csv)
```

```python
df = pd.DataFrame(columns=['rating', 'Title', 'Review'])
for i in csv:
    df1 = pd.read_csv(i)
    print("{} File shape is {}".format(i,df1.shape))
    df = df.append(df1,ignore_index=True) # Append dataframes to master dataframe df1 for csv file
    print("Final file shape",df.shape)
    print('______'*10)
```

```python
# Remove record which contain English_Reviews tag
df = df[df['Review']!='English_Reviews']
print("Dataframe shape:",df.shape)
```
